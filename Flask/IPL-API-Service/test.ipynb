```python
team1 = 'Rajasthan Royals'
team2 = 'Royal Challengers Bangalore'

filtered_df = df[((df['team1'] == team1) & (df['team2'] == team2)) |
    ((df['team1'] == team2) & (df['team2'] == team1))]
```

```python
total_matches = filtered_df.shape[0]
```

```python
matches_won_team1 = filtered_df['winner'].value_counts()[team1]
matches_won_team2 = filtered_df['winner'].value_counts()[team2]
print(matches_won_team1)
draws = total_matches - (matches_won_team1 + matches_won_team2)
```

```python
def team_vs_team(team1, team2):
    filtered_df = df[((df['team1'] == team1) & (df['team2'] == team2)) |
                ((df['team1'] == team2) & (df['team2'] == team1))]
    total_matches = filtered_df.shape[0]
    matches_won_team1 = filtered_df['winner'].value_counts()[team1]
    matches_won_team2 = filtered_df['winner'].value_counts()[team2]
    draws = total_matches - (matches_won_team1 + matches_won_team2)

    response = {
        'Total Matches' : total_matches,
        'Team 1': matches_won_team1,
        'Team 2': matches_won_team2,
        'Draw': draws
    }

    return response
```

```python
team_vs_team('Mumbai Indians', 'Royal Challengers Bangalore')
```
