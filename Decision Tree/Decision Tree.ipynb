# Decision Tree Claasifier and Regressor

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder
```

## Insurance data

```python
## Importing the Data
df1 = pd.read_csv(r"C:\Users\singh\OneDrive\Desktop\Python\Data\insurance - insurance.csv")
df1.head()
```

```python
## Encoding Process
lb = LabelEncoder()
df1['sex'] = lb.fit_transform(df1['sex'])
df1['smoker'] = lb.fit_transform(df1['smoker'])
df1['region'] = lb.fit_transform(df1['region'])

df1.head()
```

```python
## Dividing the Data
x1 = df1.drop(columns= ['charges'])
y1 = df1['charges']
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python
## Applying the ML Model
dtr = DecisionTreeRegressor()
dtr.fit(x_train1, y_train1)
```

```python
## Predicting the data
y_pred1 = dtr.predict(x_test1)
```

```python
## Evaluating the Accuracy
score1 = r2_score(y_test1, y_pred1)
print(score1)
```

## Covid data

```python
df2 = pd.read_csv(r"C:\Users\singh\OneDrive\Desktop\Python\Data\covid_toy.csv")
df2.head()
```

## Churn Modeling

```python
## Importing Dataset
df = pd.read_csv(r"C:\Users\singh\OneDrive\Desktop\Python\Data\Churn_Modelling - Churn_Modelling.csv")
df.head()
```

```python
## Dropping 
df.drop(columns= ['Surname'], inplace=True)
```

```python
## Encoding
lb = LabelEncoder()
df['Geography'] = lb.fit_transform(df['Geography'])
df['Gender'] = lb.fit_transform(df['Gender'])
```

```python
## Evaluating Accuracy
y_pred = dtc.predict(x_test)

score = accuracy_score(y_test, y_pred)
print(score)
```

## Cara Dataset

```python
df = pd.read_csv(r"C:\Users\singh\OneDrive\Desktop\Python\Data\cars - cars.csv")
df.head()
```

```python
df.drop(columns=['name'], inplace=True)
```

```python
df.info()
```

```python
## Encoding
lb = LabelEncoder()
df['fuel'] = lb.fit_transform(df['fuel'])
df['seller_type'] = lb.fit_transform(df['seller_type'])
df['transmission'] = lb.fit_transform(df['transmission'])
df['owner'] = lb.fit_transform(df['owner'])

df.head()
```

```python
## Dividing the Data
x = df.drop(columns=['km_driven'])
y = df['km_driven']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
dtr = DecisionTreeRegressor()

dtr.fit(x_train, y_train)
```

```python
y_pred = dtr.predict(x_test)

score = r2_score(y_test, y_pred)
print(score)
```
