### Nominal/OHE Encoding

One Hot Encoding is a technique used to represent categorical data as a numerical data, which is more suitable for machine learning algorithms. In this technique, each category is represented as a binary vector where each bit corressponds to a unique category.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
## Sparse Matrix --->  
```

```python
## Creating a DataFrame
df = pd.DataFrame({
    'color' : ['red', 'blue', 'green', 'green', 'red', 'blue']
})
```

```python
df.head()
```

```python
## Create an Instance of OneHotEncoder
ohe = OneHotEncoder()
```

```python
## Performing Fit and Transform
df_ohe = ohe.fit_transform(df[['color']]).toarray()
```

```python
df_new = pd.DataFrame(df_ohe, columns= ohe.get_feature_names_out())
```

```python
df_new
```

```python
pd.concat([df, df_new], axis = 1)
```
