### 38. What is the difference between rand(), randn(), and randint()?

```python
# Using rand(): Random floats between 0 and 1
rand_array = np.random.rand(2, 3)
print("rand():\n", rand_array)

# Using randn(): Random floats from standard normal distribution
randn_array = np.random.randn(2, 3)
print("\nrandn():\n", randn_array)

# Using randint(): Random integers within a specified range
randint_array = np.random.randint(10, 50, size=(2, 3))
print("\nrandint():\n", randint_array)
```

### 42. How do you vectorize a Python function using np.vectorize()?

```python
# Normal Python function (works on single values)
def square_and_check(x):
    if x % 2 == 0:
        return x**2   # square even numbers
    else:
        return x**3   # cube odd numbers

# Vectorize the function
vec_func = np.vectorize(square_and_check)

# Apply on array
arr = np.array([1, 2, 3, 4, 5])
result = vec_func(arr)

print("Original Array:", arr)
print("Vectorized Output:", result)
```

### 43. What is the difference between vectorization and loops in terms of speed?

```python
# Vectorization uses optimized C-based NumPy operations that run on entire arrays at once, making them significantly faster than 
# traditional Python loops, which process elements one-by-one. 
# Vectorized operations reduce Python-level overhead, improve memory access patterns, and take advantage of SIMD and low-level optimizations. 
# Therefore, vectorization is preferred for speed, efficiency, and cleaner code.
```
